```python
next_df["best quality"]= [1 if x>=7 else 0 for x in df.quality]

print(next_df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df=next_df.drop('quality',axis=1)
df
```

```python
df['best quality'].value_counts()
```

```python
x=df.drop('best quality',axis=1)
```

```python
y = df['best quality']
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.20,random_state=40)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
norm= MinMaxScaler()

norm_fit= norm.fit(x_train)
norm_fit
```

```python
new_xtrain= norm_fit.transform(x_train)
new_xtrain
```

```python
new_xtest = norm_fit.transform(x_test)
new_xtest
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
rnd= RandomForestClassifier()
```

```python
fit_rnd= rnd.fit(new_xtrain,y_train)
fit_rnd
```

```python
rnd_score= rnd.score(new_xtest,y_test)
rnd_score
```

```python
x_predict= rnd.fit(new_xtrain,y_train)
x_predict
```

```python
x_predict = list(rnd.predict(x_test))
predicted_df = {'predicted_values': x_predict, 'original_values': y_test}
#creating new dataframe
pd.DataFrame(predicted_df).tail(20)
```
