```python
from sklearn.feature_selection import SelectKBest,f_classif
```

```python
x=data.drop('Attrition',axis=1)
y=data.Attrition
```

```python
best_features=SelectKBest(score_func=f_classif,k=24)
```

```python
fit=best_features.fit(x,y)
```

```python
df_score=pd.DataFrame(fit.scores_)
df_columns=pd.DataFrame(x.columns)
feature_scores=pd.concat([df_columns,df_score],axis=1)
feature_scores.columns=['FeatureName','Score']
print(feature_scores.nlargest(24,'Score'))
```

```python
new_x=data[['OverTime','TotalWorkingYears','JobLevel','MaritalStatus','YearsInCurrentRole','Age','YearsWithCurrManager','StockOptionLevel','YearsAtCompany','JobInvolvement','JobSatisfaction','EnvironmentSatisfaction','DistanceFromHome','JobRole','Department','WorkLifeBalance','TrainingTimesLastYear','DailyRate','RelationshipSatisfaction','NumCompaniesWorked','YearsSinceLastPromotion','Education','Gender','EducationField']]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_scalar=scaler.fit_transform(new_x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,classification_report
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scalar,y,test_size=0.25,random_state=240)
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
cfm=confusion_matrix(y_test,y_pred)
```

```python
cfm
```

```python
print(classification_report(y_test,y_pred,digits=2))
```
